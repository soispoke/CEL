### I. Base Fees <a class="anchor" id="second-bullet"></a>

##### "Gas consumption is a cost that should be shouldered by the whole network, since every node on the network has to spend storage and computational resources to validate each message and maintain a consistent state of the network. As such, some amount of gas is burned to compensate for the network based on the gas usage of a particular message."

**1) GasUsage multiplied by BaseFee is burned as on-chain resources are consumed, and reflect how much demand there is for blockspace. On the given time period, we show that it follows a multimodal distribution, with long tail and a peak towards values close to zero, which intuitively makes sense because gas fees will converge/go back to zero when demand is low.**

**3) We also noticed an oscillation pattern in base fees over time, so we asked whether base fees followed a consistent pattern with higher fees during week days than on the weekend.**

First, we create a new "weekend" boolean column (**True** if Saturday or Sunday, **False** if Monday, Tuesday, Wednesday, Thursday and Friday).

```python
df['weekend'] =  df['timestamp'].dt.dayofweek > 4
weekend = df['mean_base_fee_burn'][df['weekend']]
weekday = df['mean_base_fee_burn'][~df['weekend']]
df[['mean_base_fee_burn','weekend']].groupby("weekend").agg("mean")
```

Since "weekday" and "weekend"  conditions were significantly unbalanced (i.e., larger number of "weekday" samples), we randomly selected conscious samples in order to match the number of "weekend" samples before performing the statistical analysis and repeated this procedure 100 times, with different segments randomly chosen every time, and observed pvalues consistently being inferior to 0.05 across repetitions.

```python
tvalue, pvalue = [],[]
for i in range(1000):
    tvalue_temp, pvalue_temp = ttest_rel(weekday.sample(n=len(weekend)),weekend)
    tvalue.append(tvalue_temp), pvalue.append(pvalue_temp)
```

**We found that the amount of base fees burned is significantly higher on week days, indicating an non negligible participation from professional actors in the Filecoin ecosystem.**

```python
# Plot Based Fees Burned over time
data = df[["mean_base_fee_burn"]]/1e18
xticks = [840,9478,19557,28917]
xticklabels = ['Tue. 06-01','Mon. 06-07','Mon. 06-14','Mon. 06-21']
xlabel_1, xlabel_2 = 'Date (8 am EST)','Repetitions'
ylabel_1, ylabel_2 = 'Base Fee Burned in attoFIL','P-value'
set_title_1, set_title_2 = 'Mean Base Fee Burned over time', 'P-values across repetitions'

fig, axs = plt.subplots(1, 2)
for i in range(2):
    if i==0:
        axs[i].set_xticks(xticks)
        axs[i].set_xticklabels(xticklabels)
        axs[i].set(xlabel=xlabel, ylabel=ylabel_1)
        axs[i].plot(data, color = 'dodgerblue',label='BaseFees Burn') 
        axs[i].plot((data).apply(savgol_filter,  window_length=1439, polyorder=2), color = 'k')
        axs[i].set_title(set_title_1)
        axs[i].legend()    
        for ii in xticks:
            if ii == 840:
                rectangle = patches.Rectangle((ii, 0),6598, 0.2, facecolor="green", linewidth=7,alpha=0.2)
            else:
                rectangle = patches.Rectangle((ii, 0),7200, 0.2, facecolor="green", linewidth=7,alpha=0.2)
            axs[i].add_patch(rectangle)
    elif i==1:
        axs[i].plot(pvalue, color = 'k')
        axs[i].set_title(set_title_2)
        axs[i].set(xlabel=xlabel_2, ylabel=ylabel_2)
fig.set_figheight(6)
fig.set_figwidth(14)

hide_code()
```

### II. Gas Limit and Overestimating Gas Usage <a class="anchor" id="third-bullet"></a>

Next, we focus on a particularly tricky property of the Filecoin protocol: it does not know how much gas a message will exactly consume ahead of execution. This means users have to estimate how much gas will be used for their transaction, by setting the GasLimit parameter.

By how gas ressources are distributed over time, we show that difference between the gas limit set by users and the gas actually used to consume blockchain ressources is split between a part that is burned. The two peaks in gas limits that probably originate from errors on the user end capture the importance not to overestimate the gas limit too much since part of the overestimated fees are burned. We also see that the remaining fees are refunded to the user.

In practice, we can see that users usually set a gas limit around 20% higher that the gas that's actually used to consume blockchain ressources. Out of the 18.8% difference called the overestimation, 77% % is actually refunded to the user, and only 4.3% is lost by the user and burned (indirectly redistributing value for all network participants by reducing the circulating supply).

```python
print('Average Diff. Between Gas Used and Gas Limit:',
      100 - np.mean(df['mean_gas_used'] * 100 / df['mean_gas_limit'])
      , '%')

select_columns = ['mean_gas_limit','mean_gas_used','mean_gas_refund','mean_gas_burned']
data = df[select_columns]
labels = ['Gas Limit','Gas Used','Gas Refund','Gas Burned']
xticks = [840,9478,19557,28917]
xticklabels = ['Tue. 06-01','Mon. 06-07','Mon. 06-14','Mon. 06-21']
xlabel_1, xlabel_2 = 'Date (8 am EST)',''
ylabel_1, ylabel_2 = 'Gas ressources in units',''
set_title_1, set_title_2 = 'Mean Gas ressources over time', 'Overestimation repartition'

fig, axs = plt.subplots(1, 2)
for i in range(2):
    if i==0:
        axs[i].set_xticks(xticks)
        axs[i].set_xticklabels(xticklabels)
        axs[i].set(xlabel=xlabel, ylabel=ylabel_1)
        
        for j,k in zip(labels,select_columns):
            axs[0].plot(data[k],label=j) 
        axs[i].set_title(set_title_1)
        axs[i].legend()    
    elif i==1:
        axs[i].pie((4.307887136953769, 14.57835145630367),labels=['Burned','Refunded'],colors=['r','g'],autopct='%1.0f%%')
        axs[i].set_title(set_title_2)
        axs[i].set(xlabel=xlabel_2, ylabel=ylabel_2)
fig.set_figheight(6)
fig.set_figwidth(14)

hide_code()
```

### III. Miners' strategies <a class="anchor" id="fourth-bullet"></a>

Currently, the default implementation is that miners select messages based on GasFeeCap/GasLimit to maximize their expected return given the GasLimit of a block. When the network is congested and BaseFee is high, miners can choose to underpack blocks, hence reducing total GasLimit, to bring down the BaseFee, but potentially at the expense of some GasPremium.

So what do miners actually do? (or at least, how did they behave during the month of June 2021)

To find out, we checked if BaseFee is correlated with GasPremium. We grouped and averaged BaseFee and GasPremium per hour, and used Spearman's rank correlation to show that BaseFee showed a significant positive correlation (Rcoef = 0.67, p < 0.01) with GasPremium. The results are displayed in the figure below:

```python
base_fee = df[['timestamp','mean_base_fee']].groupby(pd.Grouper(key='timestamp', axis=0, 
              freq='H')).mean().values.flatten()
gas_premium = df[['timestamp','mean_gas_premium']].groupby(pd.Grouper(key='timestamp', axis=0, 
              freq='H')).mean().values.flatten()

xticks = [840,9478,19557,28917]
xticklabels = ['Tue. 06-01','Mon. 06-07','Mon. 06-14','Mon. 06-21']
xlabel_1, xlabel_2 = 'Date (8 am EST)','Mean Base Fees (attoFIL/Gas Unit)'
ylabel_1, ylabel_2 = 'Mean Base Fees (attoFIL/Gas Unit)','Mean Gas Premium (attoFIL/Gas Unit)'
set_title_1, set_title_2 = 'Mean Base Fees (attoFIL/Gas Unit)', 'Correlation between Base Fees and Gas Premium'

fig, axs = plt.subplots(1, 2)
for i in range(2):
    if i==0:
        ax2 = axs[i].twinx()
        ax2.plot(gas_premium, color = 'orange',label='Gas Premium')

        ax2.set_ylabel('Gas Premium (attoFIL/Gas Unit)',labelpad=10,rotation=-90)
        ax2.legend(loc=9)
        axs[i].set_xticks(xticks)
        axs[i].set_xticklabels(xticklabels)
        axs[i].set(xlabel=xlabel, ylabel=ylabel_1)
        axs[i].plot(base_fee, color = 'dodgerblue',label='Base Fees')
        axs[i].set_title(set_title_1)
        axs[i].legend()    
    elif i==1:
        axs[i].scatter(base_fee, gas_premium, s=5, color = 'k')
        axs[i].plot(np.unique(base_fee), np.poly1d(np.polyfit(base_fee, gas_premium, 1))(np.unique(base_fee)),color='r')
        axs[i].set_title(set_title_2)
        axs[i].set(xlabel=xlabel_2, ylabel=ylabel_2)
fig.set_figheight(8)
fig.set_figwidth(18)
print(spearmanr(base_fee,gas_premium))
hide_code()
```
